```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

link = "/content/drive/MyDrive/DATASET/smoking.csv"
dataset = pd.read_csv(link)

print(dataset)
```

```python
print("Statistik Fitur Dataset smoking:")
print(dataset.describe())
```

```python
import matplotlib.pyplot as plt

# Membuat Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot([dataset[dataset['amt_weekends'] == 0]['age'].dropna(),
             dataset[dataset['amt_weekdays'] == 1]['age'].dropna()],
            labels=['Not Survived', 'Survived'])
plt.title('Box Plot of Age by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


# Membuat Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(dataset['age'], dataset['amt_weekdays'], c=dataset['amt_weekends'], cmap='coolwarm', alpha=0.6)
plt.title('Scatter Plot of Age vs Fare (Survival Status)')
plt.xlabel('Age')
plt.ylabel('gros')
plt.colorbar(label='Survived')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = dataset.corr()

# Membuat Heatmap
plt.figure(figsize=(20, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
