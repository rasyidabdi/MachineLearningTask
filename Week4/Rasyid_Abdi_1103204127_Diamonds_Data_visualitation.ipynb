```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/DATASET/diamonds.csv')

# Memilih hanya fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64'])

# Membuat box plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_features.columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df[col], color='pink')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# @title Distribution Plot untuk setiap fitur numerik
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
diamonds = pd.read_csv('/content/drive/MyDrive/DATASET/diamonds.csv')

# Select numeric features
numeric_features = diamonds.select_dtypes(include=['float64', 'int64'])

# Plot Distribution Plot for each numeric feature
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_features.columns):
    plt.subplot(3, 3, i+1)
    sns.histplot(diamonds[column], kde=True, color='skyblue')
    plt.title(column)
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
# @title Violin Plot untuk setiap fitur numerik
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
diamonds = pd.read_csv('/content/drive/MyDrive/DATASET/diamonds.csv')

# Select numeric features
numeric_features = diamonds.select_dtypes(include=['float64', 'int64'])

# Plot Violin Plot for each numeric feature
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_features.columns):
    plt.subplot(3, 3, i+1)
    sns.violinplot(x=diamonds[column], orient='v', palette='Set2')
    plt.title(column)
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
# @title Heat Map
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/melb_data.csv')

# Memilih hanya fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
